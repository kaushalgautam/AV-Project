```python
df1 = df
```

```python
criteria = df1["Manager_Business"] < 0
df1["Manager_Business"] = df1["Manager_Business"].abs()
df1["Manager_Business_Negative"] = criteria
```

```python
criteria = df1["Manager_Business2"] < 0
criteria.value_counts()
df1["Manager_Business2"] = df1["Manager_Business2"].abs()
df1["Manager_Business_Negative2"] = criteria
```

```python
df1.head()
```

```python
df1.describe()
```

```python
df1.shape
```

```python
import datetime
a_birthdates = pd.to_datetime(df1["Applicant_BirthDate"])
a_application_dates = pd.to_datetime(df1["Application_Receipt_Date"])
applicant_age = a_application_dates - a_birthdates
```

```python
ages = []
for i in range(0, applicant_age.size):
    ages.append(applicant_age[i].days)
```

```python
df1["Applicant_Age_Initially"] = ages
```

```python
m_birthdates = pd.to_datetime(df1["Manager_DoB"])
m_application_dates = pd.to_datetime(df1["Application_Receipt_Date"])
manager_age = m_application_dates - m_birthdates
ages = []
for i in range(0, manager_age.size):
    ages.append(manager_age[i].days)
#df1["Applicant_Age_Initially"] = ages
df1["Manager_Age"] = ages
```

```python
m_doj =pd.to_datetime(df1["Manager_DOJ"])
manager_tenure = m_application_dates - m_doj
tenures = []
for i in range(0, manager_tenure.size):
    tenures.append(manager_tenure[i].days)
df1["Manager_Tenure"] = tenures
```
